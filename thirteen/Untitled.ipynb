```python
import pandas as pd
from statsmodels.tsa.stattools import adfuller

data=pd.read_excel('D:\M\MATLAB Driver/forward/thirteen/山东贸易.xlsx',sheet_name='Sheet1')
# 定义显著性水平
alpha = 0.05

# 创建一个空的DataFrame来存储结果
result_df = pd.DataFrame(columns=["ADF Statistic", "p-value", "Critical Values", "Stationary"])

# 对每一列进行ADF检验，并将结果存储在DataFrame中
for col in data.columns:
    result = adfuller(data[col])
    stationary = "Yes" if result[1] < alpha else "No"
    result_df.loc[col] = [result[0], result[1], result[4], stationary]

# 打印结
result_df
```

```python
#导出excel
result_df.to_excel('D:\M\MATLAB Driver/forward/thirteen/ADF检验.xlsx')
```
