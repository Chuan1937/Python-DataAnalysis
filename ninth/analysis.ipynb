# 二 影响因素

对疼痛知识的认知情况及影响因素中的“影响因素”侧重点在于影响老人认知疼痛的外在与内在因素，例如经济，心理，社会状况等。即问卷中的数据能从侧面反应问题。

```python
# 疼痛对健康有影响	您知道怎样正确使用止痛药	止痛药能够控制疼痛	人们很容易对止痛药上瘾	与止痛药的副作用相比，忍耐疼痛更加容易	止痛药应该在疼痛剧烈时使用
#将这几列进行组合成一个新的dataframe，命名为data1
data1=data[['疼痛对健康有影响','您知道怎样正确使用止痛药','止痛药能够控制疼痛','人们很容易对止痛药上瘾','与止痛药的副作用相比，忍耐疼痛更加容易','止痛药应该在疼痛剧烈时使用']]
```

```python
#计算data1的相关系数
data1.corr()
```

这几种变量的相关性都不强，但是可以看出，在使用止痛药副作用与忍耐疼痛相比下，更倾向于忍耐疼痛。与此情况下所反应的是使用疼痛药的耗资情况与老人的讳疾忌医心理。

```python
#您会感到疼痛很神秘	您认为疼痛不可解除	您对自身的疼痛怀有自责感	您认为自己需要更多关于疼痛知识的信息吗？
#将这几列进行组合成一个新的dataframe，命名为data2
data2=data[['您对自身的疼痛怀有自责感','您认为疼痛不可解除','您会感到疼痛很神秘']]
#计算data2的相关系数
data2.corr()
```

同样，对疼痛的理解不到位，畏惧疼痛也影响了疼痛认知缺失的情况。

#经济情况应该是最能影响疼痛认知的因素之一，大部分老人的经济来源是退休金，而面对医院的高昂费用，很多老人会选择忍受疼痛，而不去医院就医。强行忍耐。

```python
#您认为自己需要哪些形式的信息和资源来了解疼痛知识（多选）？
#将这一列进行分割，按┋进行分割
t2=data['您认为自己需要哪些形式的信息和资源来了解疼痛知识（多选）？'].str.split('┋')
#将t2转换成dataframe
t2=pd.DataFrame(t2)
#将“其他”替换为0
t2=t2.replace('其他',0)
#对t2每行进行分别统计字符个数，其他替换为0，其他的就是字符个数
t2=t2.applymap(lambda x: len(x) if x!=0 else 0)
#绘图t2柱状图
sns.countplot(y='您是否参加过跟疼痛管理有关的社区活动或组织？',hue=t2['您认为自己需要哪些形式的信息和资源来了解疼痛知识（多选）？'],data=data)
```

经常参加活动的人很少需要其他形式来了解疼痛知识，而从未参加的却十分渴望了解疼痛知识，这可能是因为经常参加活动的人已经了解了疼痛知识，并且更能够身临其境。而未参加的可能是因为身体原因，或者是思想观念等，不愿意参加活动，害怕上当。

```python
data_new=data[['您认为自己需要更多关于疼痛知识的信息吗？','您的年龄：','您的性别：','您的文化程度：','您是否接受过疼痛知识宣教：','您目前身体健康程度：']]
data_new
```

```python
#将1.非常需要，2.需要，3.不需要，4.非常不需要，转换成1，2，3，4
data_new=data_new.replace('1.非常需要',0)
data_new=data_new.replace('2.需要',1)
data_new=data_new.replace('3.不确定',2)
data_new=data_new.replace('4.不需要',3)
#将性别转换成0，1
data_new=data_new.replace('男',0)
data_new=data_new.replace('女',1)
#将文化程度的中专及高中，初中及以下，大专及以上转换成0，1，2
data_new=data_new.replace('中专及高中',0)
data_new=data_new.replace('初中及以下',1)
data_new=data_new.replace('大专及以上',2)
#将是否接受过疼痛知识宣教转换成0，1
data_new=data_new.replace('是',0)
data_new=data_new.replace('否',1)
#将身体健康程度转换成身体健康，身体患有某种或多种疾病转换成0，1
data_new=data_new.replace('身体健康',0)
data_new=data_new.replace('身体患有某种或多种疾病',1)

data_new
```

```python
#标准化
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(data_new)
data_new=scaler.transform(data_new)
data_new=pd.DataFrame(data_new)
```

```python
#将data_new的列名改成原来的列名
data_new.columns=['您认为自己需要更多关于疼痛知识的信息吗？','您的年龄：','您的性别：','您的文化程度：','您是否接受过疼痛知识宣教：','您目前身体健康程度：']
#第一列用原来的值替换
data_new['您认为自己需要更多关于疼痛知识的信息吗？']=data['您认为自己需要更多关于疼痛知识的信息吗？']
```
